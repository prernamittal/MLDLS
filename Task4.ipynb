```python
#importing libraries
#Keras is a neural network Application Programming Interface (API) for Python that is tightly integrated with TensorFlow, which is used to build machine learning models.
import tensorflow as tf
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Salary_Data.csv')
```

```python
#creating array of x and y such that y=(x/2) - 4
#A numpy array is a grid of values, all of the same type, and is indexed by a tuple of nonnegative integers.
Xs= df.iloc[:,0].values
Ys= df.iloc[:,1].values
```

```python
#defining a single layer; single neuron model
model = tf.keras.Sequential([keras.layers.Dense(units=3, activation='relu', input_shape=[1])])
#Sequential groups a linear stack of layers into a tf.keras
model.add(Dense(1))
#Dense Layer is simple layer of neurons in which each neuron receives input from all the neurons of previous layer.
print(model.summary())
```

```python
#Converts a Keras model to dot format and save to a file.
tf.keras.utils.plot_model(
    model,
    to_file="model.png",
    show_shapes=True,
    show_dtype=False,
    show_layer_names=True,
    rankdir="TB",
    expand_nested=False,
    dpi=96,
)
```

```python
#defining the model evaluation for backpropogation
#Compile defines the loss function, the optimizer and the metrics.
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
#taraining the model with iterations (No. of front & back propogation)
run_model = model.fit(Xs, Ys, epochs=700)
```

```python
print(model.predict([40.0]))
#predicting the value of Ys as 40 is our Xs
#Output is 15.25 for epochs 500 and 15.99 for epochs 700 
#actual output should have been 16
```

```python
#list all data in history
print(run_model.history.keys())
```

```python
# summarize history for accuracy
plt.plot(run_model.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()
```
